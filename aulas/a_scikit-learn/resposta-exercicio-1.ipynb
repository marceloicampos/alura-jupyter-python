```python
from sklearn import datasets, model_selection, svm, metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X_blob, y_blob = datasets.make_blobs(n_samples=2000, n_features=2, centers=4, random_state=32)
```

```python
plt.scatter(X_blob[:, 0], X_blob[:, 1], c=y_blob, s=15);
```

```python
X_train, X_test, y_train, y_test = \
    model_selection.train_test_split(X_blob, y_blob, test_size=.3, random_state=42)
```

```python
estimator = svm.SVC(kernel="linear")
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.score(X_test, y_test)
```

```python
y_pred = estimator.predict(X_test)
metrics.confusion_matrix(y_test, y_pred)
```
